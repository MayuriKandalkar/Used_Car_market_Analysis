## Steps in Data Analysis Using Python:

1. Import neccesary Libraries like numpy, pandas, matplotlib, seaborn , etc
2. Load the Dataset
3. Data Understanding (checknull values, remove duplicates (if any) , check data type , rename columns, etc.
4. Data Reduction ( Removes Column that are not part of analysis)
5. Feature Engineering ( Create new columns by manipulating existing ones or manipulate existing columns itself)
6. Data Wrangling/Data Cleaning
      - Data Wrangling means checking data in values level or checking the vlues in rows.
7. Exploratory Data Analystics (EDA):
   - Univariate  Analysis
       - one variable means one column (avg age or avg marks)
   - Biariate Analysis
       - scatter plot is bivariate
   - Multivariate

```python
# milage column logic

def kmpl_kmkg_converter(mileage):
    if "km/kg" in mileage:
        value = float(mileage.replace(" km/kg", ""))
        return value * 0.75
    elif "kmpl" in mileage:
        value = float(mileage.replace(" kmpl", ""))
        return value
    else:
        return np.nan
```

```python
kmpl_kmkg_converter("25.6 km/kg")
```

```python
kmpl_kmkg_converter("25.6 kmpl")
```

```python
df["Mileage"] = df["Mileage"].apply(lambda x: kmpl_kmkg_converter(str(x)))
df.info()
```

# Budged Car Analysis

```python
budget_df = df[df['Price'] < 10]
```

```python
budget_df
```

```python
cat_col = budget_df.select_dtypes(include = ["object"]).columns.tolist()
cat_col
```

```python
num_col = budget_df.select_dtypes(include = ["int", "float"]).columns.tolist()
num_col
```

```python
for col in cat_col:
    print(col)
    sns.countplot(budget_df[col], order = budget_df[col].value_counts().index)
    plt.show()
```

## Univariate Analysis for Budget Cars

- **City-wise Distribution:** Mumbai leads the largest market for used cars, followed by Hyderabad and Pune. Ahmedabad has the smallest market presence.
- **Fuel Type:** Petrol and diesel vehicles dominate the used car segment, while CNG and LPG options represent a small fraction of the market.
- **Transmission Type:** Manual transmission vehicles are significantly more prevalent in the budget used car category.
- **Ownership:** A majority of the listed used cars are first-owner vehicles, indicating relatively new ownership history.
- **Seating Capacity:** Most budget used cars are 5-seaters, with 7-seaters being the next most common configuration.
- **Brands Popularity:** Maruti leads market, followed by Hyundai and Honda.



```python
for col in num_col:
    print(col)
    plt.figure(figsize = (16,6))
    plt.subplot(1,2,1)
    budget_df[col].hist(grid = False)
    plt.ylabel("Count")
    plt.subplot(1,2,2)
    sns.boxplot(x=budget_df[col])
    plt.show()
```

```python

# Removal of Outlaiers 
```

```python
for col in num_col:
    Q1 = budget_df[col].quantile(0.25)
    Q3 = budget_df[col].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 *IQR
    budget_df = budget_df[(budget_df[col]>= lower_bound) & (budget_df[col] <=upper_bound)]
```

```python
for col in num_col:
    print(col)
    plt.figure(figsize = (16,6))
    plt.subplot(1,2,1)
    budget_df[col].hist(grid = False)
    plt.ylabel("Count")
    plt.subplot(1,2,2)
    sns.boxplot(x=budget_df[col])
    plt.show()
```

# Bivariate EDA:

```python
plt.figure(figsize=(12,8))
sns.pairplot(data = budget_df)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=budget_df, y="Price", x="Fuel_Type")
plt.show()
```

## Multivariate EDA:

```python
num_col = budget_df.select_dtypes(include=["number"])
num_col
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(num_col.corr(), annot= True, fmt= ".2f")
plt.show()
```

This is a correlation heatmap showing how different features of cars are related to each other.car age shows a strong negative correlation(inverse) with price. This means that as a car gets older, its price tends to decrease significantly.
